```python
# import numpy and pyplot and get a list of months and items sold
import numpy as np
import matplotlib.pyplot as plt

months = list(['January', 'February', 'March', 'April', 'May', 'June'])
items_sold = np.linspace(1000, 2000, 6)
```

```python
# create a line plot
plt.title('Items Sold has been Increasing Linearly')
plt.xlabel('Month')
plt.ylabel('Items Sold')
plt.plot(months, items_sold, '*:')

plt.show()
```
