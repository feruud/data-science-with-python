```python
import pandas as pd
import numpy as np

items_sold_df = pd.read_csv('../data/Items_Sold_by_Week.csv')
weight_by_height = pd.read_csv('../data/Weight_by_Height.csv')
normal_numbers = np.random.normal(loc = 0, scale = 0.1, size = 100)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows = 3, ncols = 2, figsize = (12,12))
```

```python
axes[0, 0].plot(items_sold_df.Week, items_sold_df.Items_Sold)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('Week')
axes[0, 0].set_ylabel('Items Sold')

axes[0, 1].bar(items_sold_df.Week, items_sold_df.Items_Sold)
axes[0, 1].set_title('Bar Plot')
axes[0, 1].set_xlabel('Week')
axes[0, 1].set_ylabel('Items Sold')

axes[1, 0].barh(items_sold_df.Week, items_sold_df.Items_Sold)
axes[1, 0].set_title('Horizontal Bar Plot')
axes[1, 0].set_xlabel('Week')
axes[1, 0].set_ylabel('Items Sold')

axes[1, 1].hist(normal_numbers, bins = 25)
axes[1, 1].set_title('Histogram Plot')
axes[1, 1].set_xlabel('Numbers')
axes[1, 1].set_ylabel('Frequency')

axes[2, 0].scatter(x = weight_by_height.Weight, y = weight_by_height.Height)
axes[2, 0].set_title('Scatter Plot')
axes[2, 0].set_xlabel('Weight')
axes[2, 0].set_ylabel('Height')

axes[2, 1].boxplot(normal_numbers)
axes[2, 0].set_title('Box And Whisker Plot')

plt.tight_layout()

fig.savefig('Activity 4')

fig

```
