```python
# import numpy and get an array of 20 numbers from 0 to 10 
import numpy as np

x = np.linspace(0, 10, 20)
```

```python
# get two more arrays one with the squared numbers and one with the cubed numbers
y = x**3
y2 = x**2
```

```python
# import pyplot from matplotlib and create a line plot with two lines
import matplotlib.pyplot as plt

plt.title('x by x Cubed')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.plot(x, y, 'D-k')
plt.plot(x, y2)

plt.show()
```

```python
# change second line to be a dashed red line
plt.title('x by x Cubed')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.plot(x, y, 'D-k')
plt.plot(x, y2, '--r')

plt.show()
```

```python
# add labels for both lines
plt.title('x by x Cubed')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.plot(x, y, 'D-k', label = 'x cubed')
plt.plot(x, y2, '--r', label = 'x squared')
plt.legend(loc = 'upper left')

plt.show()
```

```python
# add a multiple line title
plt.title('As x increases\nx cubed (black) increases \nat a greater rate than\nx squared(red)')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.plot(x, y, 'D-k', label = 'x cubed')
plt.plot(x, y2, '--r', label = 'x squared')
plt.legend(loc = 'upper left')

plt.show()
```

```python
# change the plot size
plt.title('As x increases\nx cubed (black) increases \nat a greater rate than\nx squared(red)')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.plot(x, y, 'D-k', label = 'x cubed')
plt.plot(x, y2, '--r', label = 'x squared')
plt.legend(loc = 'upper left')
plt.figure(figsize = (10, 5))

plt.show()
```
