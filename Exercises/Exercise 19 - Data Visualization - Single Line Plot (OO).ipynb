```python
import numpy as np

x = np.linspace(0, 10, 20)
y = x**3
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots()
axes.plot(x, y, 'D-k')
axes.set_xlabel('Linearly spaced numbers')
axes.set_ylabel('y Value')
axes.set_title('As x increases, y increases by x cubed')
plt.show()
```
