```python
import numpy as np

x = np.linspace(0, 10, 20)
```

```python
y = x**3
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)

plt.show()
```

```python
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.title('x by x Cubed')
plt.plot(x, y, 'k')

plt.show()
```

```python
plt.plot(x, y, 'Dk')

plt.show()
```

```python
plt.plot(x, y, 'D-k')

plt.show()
```

```python
plt.plot(x, y, 'D-k')
plt.title('x by x Cubed', fontsize = 22)
plt.show()
```
